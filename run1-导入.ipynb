```python
# 名义利率 - TIPS实际利率 = 盈亏平衡通胀预期
nominal_10y = web.DataReader("DGS10", "fred", start="2020-01-01")  # 10年名义国债
tips_10y    = web.DataReader("DFII10", "fred", start="2020-01-01")  # 10年TIPS实际

breakeven = nominal_10y.join(tips_10y, lsuffix="_nominal", rsuffix="_tips")
breakeven["breakeven_inflation"] = breakeven["DGS10"] - breakeven["DFII10"]
```

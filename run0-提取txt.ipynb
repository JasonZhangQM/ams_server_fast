```python
from common.utils import extract_futures_transactions,extract_futures_deposit
from app.bills.config import Config as BlsCfg
folder_path = BlsCfg.FOLDER_QH  
```

```python
file_names = [
    f.name for f in folder_path.iterdir() 
    if f.is_file() and f.name.endswith('.txt')
    ]
for file_name in file_names:
    print(f"{file_name}...")
    file_path = folder_path / f"{file_name}"
    df = extract_futures_transactions(file_path)
    df.to_excel(folder_path / f'{file_name[:-4]}-成交记录.xlsx',index=False)
```

```python
file_names = [
    f.name for f in folder_path.iterdir() 
    if f.is_file() and f.name.endswith('.txt')
    ]
for file_name in file_names:
    print(f"{file_name}...")
    file_path = folder_path / f"{file_name}"
    df = extract_futures_deposit(file_path)
    # 增加必要字段
    df['price'] = 0.0
    df['vol'] = 0
    df['币种'] = '人民币' #期权、期货、信用
    df = df.rename(columns={'发生日期':'清算日期'}) #期货
    df['amount_act'] = df['入金']-df['出金']
    df['amount'] = abs(df['入金']-df['出金'])
    df.loc[df['amount_act']>0, 'exec_type'] = '入金' #
    df.loc[df['amount_act']<0, 'exec_type'] = '出金' #
    df.loc[df['说明'].str.contains('申报费'), 'exec_type'] = '申报费' #
    df.to_excel(folder_path / f'{file_name[:-4]}-出入金.xlsx',index=False)
```
